```python
import matplotlib.pyplot as plt

# Ranks in the Pakistan Army
ranks = {
    "Commissioned Officers": ["Second Lieutenant", "Lieutenant", "Captain", "Major",
                              "Lieutenant Colonel", "Colonel", "Brigadier", "Major General",
                              "Lieutenant General", "General", "Field Marshal"],
    "Junior Commissioned Officers": ["Naib Subedar", "Subedar", "Subedar Major"],
    "Non-Commissioned Officers and Soldiers": ["Sepoy", "Lance Naik", "Naik", "Havaldar",
                                               "Company Quartermaster Havaldar (CQMH)",
                                               "Battalion Quartermaster Havaldar (BQMH)",
                                               "Company Havaldar Major (CHM)",
                                               "Battalion Havaldar Major (BHM)"]
}

# Colors for different categories
colors = {
    "Commissioned Officers": "royalblue",
    "Junior Commissioned Officers": "forestgreen",
    "Non-Commissioned Officers and Soldiers": "firebrick"
}

# Create a figure
fig, ax = plt.subplots(figsize=(11, 8))

# Current position for the rank to be plotted
y_pos = 0

# Loop through the categories and ranks
for category, rank_list in ranks.items():
    for rank in rank_list:
        ax.scatter([category], [y_pos], color=colors[category], s=100)
        ax.text(category, y_pos, f"  {rank}", verticalalignment='center')
        y_pos += 1

# Set labels and title
ax.set_yticks([])
ax.set_xticks(list(colors.keys()))
ax.set_xticklabels(list(colors.keys()), rotation=45)
ax.set_title('Pakistan Army Ranks')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Ranks in the Pakistan Army
ranks = {
    "Commissioned Officers": ["Second Lieutenant", "Lieutenant", "Captain", "Major",
                              "Lieutenant Colonel", "Colonel", "Brigadier", "Major General",
                              "Lieutenant General", "General", "Field Marshal"],
    "Junior Commissioned Officers": ["Naib Subedar", "Subedar", "Subedar Major"],
    "Non-Commissioned Officers and Soldiers": ["Sepoy", "Lance Naik", "Naik", "Havaldar",
                                               "Company Quartermaster Havaldar (CQMH)",
                                               "Battalion Quartermaster Havaldar (BQMH)",
                                               "Company Havaldar Major (CHM)",
                                               "Battalion Havaldar Major (BHM)"]
}

# Assign a numeric level to each rank for plotting
rank_levels = {rank: i for i, category in enumerate(ranks.values(), 1) for rank in category}

# Split ranks into categories for plotting
commissioned = [rank for rank in ranks["Commissioned Officers"]]
junior_commissioned = [rank for rank in ranks["Junior Commissioned Officers"]]
non_commissioned = [rank for rank in ranks["Non-Commissioned Officers and Soldiers"]]

# Create a figure
fig, ax = plt.subplots(figsize=(15, 10))

# Plot each category of ranks
ax.barh(commissioned, [rank_levels[rank] for rank in commissioned], color='blue', label='Commissioned Officers')
ax.barh(junior_commissioned, [rank_levels[rank] for rank in junior_commissioned], color='green', label='Junior Commissioned Officers')
ax.barh(non_commissioned, [rank_levels[rank] for rank in non_commissioned], color='red', label='Non-Commissioned Officers and Soldiers')

# Set labels and title
ax.set_xlabel('Rank Level')
ax.set_title('Hierarchy of Ranks in the Pakistan Army')
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
